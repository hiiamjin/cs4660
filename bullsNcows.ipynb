```python
def bulls_and_cows1(secret: list, guess: list):
    bulls = 0
    cows = 0
    unmatched = {}
    for (s, g) in zip(secret, guess):
        if s == g:
            bulls += 1
        else:
            if unmatched.setdefault(s, 0) > 0:
                cows += 1
                unmatched[s] -= 1
            else:
                unmatched[s] = unmatched.setdefault(s, 0) - 1
            if unmatched.setdefault(g, 0) < 0:
                cows += 1
                unmatched[g] += 1
            else:
                unmatched[g] = unmatched.setdefault(g, 0) + 1
    return (bulls, cows)

# : list ... part just to make it easier to know what type it is. Can run without it
def bulls_and_cows2(secret: list, guess: list):
    bulls = 0
    cows = 0
    s_unmatched = []
    g_unmatched = []
    for (s, g) in zip(secret, guess):
        if s == g:
            bulls += 1
        else:
            s_unmatched.append(s)
            g_unmatched.append(g)
    for g in g_unmatched:
        if g in s_unmatched:
            s_unmatched.remove(g)
            cows += 1
    return (bulls, cows)


def bulls_and_cows3(secret, guess):
    bulls = 0
    s_unmatched = {}
    g_unmatched = {}
    for (s, g) in zip(secret, guess):
        if s == g:
            bulls += 1
        else:
            s_unmatched[s] = s_unmatched.get(s, 0) + 1
            g_unmatched[g] = g_unmatched.get(g, 0) + 1
    cow_dict = {k:min(v, g_unmatched[k]) for (k, v) in s_unmatched.items() if k in g_unmatched}
    cows = sum(cow_dict.values())
    return (bulls, cows)

def run_bulls_and_cows(secret, guess):
    print('\ncompare {}\n    and {}'.format(secret, guess))
    print(bulls_and_cows1(secret, guess))
    print(bulls_and_cows2(secret, guess))
    print(bulls_and_cows3(secret, guess))


run_bulls_and_cows('1038', '1234')    
run_bulls_and_cows('1038', '1380')
run_bulls_and_cows('1038', '3810')
run_bulls_and_cows('8338', '3838')
run_bulls_and_cows('3388', '8331')
run_bulls_and_cows('8338', '3833')
```
